```python
import random as rand
import math
import matplotlib.pyplot as plt
N = 1000
```

```python
def learn():
    labels = []
    features = [(rand.random() * 10 + 1, rand.random() * 10 + 1) for i in range(N)]
    for i in range(N):
        if (int(10-features[i][0])>features[i][1]) and (features[i][0]>features[i][1]):
            labels.append('2')
        elif (int(10-features[i][0])<features[i][1]) and (features[i][0]>features[i][1]):
            labels.append('1')
        elif (int(10-features[i][0])<features[i][1]) and (features[i][0]<features[i][1]):
            labels.append('4')
        elif (int(10-features[i][0])>features[i][1]) and (features[i][0]<features[i][1]):
            labels.append('3')
    print(labels)
    return features, labels
```

```python
def fDist(X1,X2):
    return math.sqrt((X1[0]-X2[0])**2 + (X1[1]-X2[1])**2)
```

```python
def most_common(lst):
    return max(set(lst), key=lst.count)
```

```python
def fClass(X,k):
    distList = [(fDist(features[i],X), labels[i]) for i in range(N)] #вызов дистанций между искомой точкой и каждой точкой фичи (вместе с лейблом каждой из них)
    distList.sort()
    answList = [distList[i] for i in range(k)]
    
    a = [answList[i][1] for i in range(len(answList))]   #взятие классов соседей с наименьшей дистанцией 
    
    return most_common(a) #просчёт класса(наибольшего кол-ва одинаковых соседей)
```

```python
def fValid(n, k):
    mis = 0
    for i in range(n):
        x = (rand.random() * 10 + 1, rand.random() * 10 + 1)
        y = fClass(x, k)        
    return y
```
